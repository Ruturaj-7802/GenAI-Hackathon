```python
import json
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.chat_models import ChatOpenAI

# Define the Prompt Template
prompt_template = PromptTemplate(
    input_variables=["instructions", "dataframe", "input_format", "output_format"],
    template="""
    Instructions:
    {instructions}
    
    Data for Recommendations:
    {dataframe}
    
    Input Format: Phases can have multiple items in team, and can have multiple items in phases too
    {input_format}

    Output Format: Skill should return list of skills, reason should be in form of texts in bullet form.
    {output_format}
    """
)

# Inputs
instructions = """
You have been given a new project with its solution, Client Industry, domain, functional area, and tools for the project. 
Create a skill matrix of resources (employees) based on their skills, rating, experience, and domain. Using the matrix, suggest 
5 resources or employees from the dataset who best fit the project requirements based on their skills, experience, rating, domain, and availability. 
Consider bandwidth as well. Provide justification for the choices. Select Employees OR Resources only from the given data. Use the columns as given in the dataset.
"""

input_format = """
Input: 
    {
        "client": "", 
        "industry": "", 
        "data_sources": ["", "", "", "", ""], 
        "skills_required": ["", "", "", ""], 
        "technologies_required": ["", "", "", "", ""], 
        "phases": [{ "duration": "", "team": [{ "role": "", "cnt": }, { "role": "", "cnt": }, { "role": "", "cnt":  }, { "role": "", "cnt":  }] }]
    }
"""

# dataset_prompt = f'The Dataset: {df_json}'

output_format = """
    Project Name: <PROJECT NAME>
        Resource Name 1: <EMPLOYEE NAME>
        Resource ID 1: <RESOURCE ID>
        Designation : <PRESENT DESIGNATION>
        Skill : <RESOURCE SKILL>
        Reason : <REASON TO SELECT THE RESOURCE>
        
        Resource Name 2: <RESOURCE NAME>
        Resource ID 2: <RESOURCE ID>
        Designation : <PRESENT DESIGNATION>
        Skill : <RESOURCE SKILL>
        Reason : <REASON TO SELECT THE RESOURCE>
        
        Resource Name 3: <RESOURCE NAME>
        Resource ID 3: <RESOURCE ID>
        Designation : <PRESENT DESIGNATION>
        Skill : <RESOURCE SKILL>
        Reason : <REASON TO SELECT THE RESOURCE>
        
        Resource Name 4: <RESOURCE NAME>
        Resource ID 4: <RESOURCE ID>
        Designation : <PRESENT DESIGNATION>
        Skill : <RESOURCE SKILL>
        Reason : <REASON TO SELECT THE RESOURCE>
        
        Resource Name 5: <RESOURCE NAME>
        Resource ID 5: <RESOURCE ID>
        Designation : <PRESENT DESIGNATION>
        Skill : <RESOURCE SKILL>
        Reason : <REASON TO SELECT THE RESOURCE>    
"""


# Combine Everything
prompt = prompt_template.format(
    instructions=instructions,
    dataframe=df_json,
    input_format=input_format,
    output_format=output_format
)

# Initialize the LLM with ChatOpenAI
# openai_key = "your_openai_api_key_here"  # Replace with your actual OpenAI API key
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0, openai_api_key=openai_key)

# Create the LLM Chain
llm_chain = LLMChain(llm=llm, prompt=prompt_template)

# Run the LLMChain with the inputs using invoke
response = llm_chain.invoke({
    "instructions": instructions,
    "dataframe": df_json,
    "input_format": input_format,
    "output_format": output_format
})

# Print the Response
print(response)
```

```python
def chunk_data(data, chunk_size=3):
    """Splits the data into smaller chunks of the specified size."""
    for i in range(0, len(data), chunk_size):
        yield data[i:i + chunk_size]


chunked_data = list(chunk_data(df_json, chunk_size=2))  # Adjust chunk_size based on the LLM's context limit

shortlisted_employees = []

for chunk in chunked_data:
    chunk_as_json = json.dumps(chunk, indent=2)
    response = shortlist_chain.invoke({
        "dataframe": chunk_as_json,
        "skills_required": skills_required,
        "technologies_required": technologies_required
    })
    shortlisted_employees.extend(response["shortlisted_employees"])  # Collect results

shortlisted_employees_as_json = json.dumps(shortlisted_employees, indent=2)
scored_response = scoring_chain.invoke({
    "shortlisted_employees": shortlisted_employees_as_json
})
```
