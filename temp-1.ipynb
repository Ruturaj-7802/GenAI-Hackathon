```python
def create_prompt(data_json, usr_pmt, usr_fmt):
    # Construct the prompt as a template
    prompt = f"""
    Role: Project Staffing Lead at a Client Company
    
    I have the following data on resources and projects in JSON format: 
    {json.dumps(data_json, indent=2)}

    The main columns are explained as follows:
    Project ID:Primary Key for project
    Project Start Date:Start date of the project
    Project End Data:End date of the project
    Company Name:Name of the client
    Project Name:Name of the project
    Description:Description of project
    Team Composition:Size of the team in bandwidth numbers
    Client Industry:Industry name for the client company (e.g., SAAS, Retail, etc.)
    Client Sub Industry:Sub industry name for the client company
    Solution Domain:Domain of project (e.g., ARR Analysis, Data Management, Accounting, etc.)
    Functional Area:Scope of the project in terms of functional area (e.g., Budget Modelling, Sales Performance Analysis, etc.)
    Tool:Tools used in the project (e.g., Azure, Excel, SQL, etc.)
    Rating From Client:Rating provided from the client on overall project (Scale of 1 to 5)
    Project Type:Type of engagement (e.g., Retainer, Pilot, Hybrid)
    Resource ID:Primary key for resource
    Job Role:Job role resource played in the project (e.g., Analyst, Manager, Data Engineer, etc.)
    Bandwidth:Bandwidth of resource in project (in percentage)
    Resource Start Date:Resource staffing start date in project
    Resource End Date:Resource staffing end date in project
    Resource Name:Name of the resource
    Present Designation:Present designation of resource
    Leaves in Next 2 weeks:Availablity of the resource (Yes/No)
    Resource SkillSets:Skillsets of the resource
    Certifications:Certification of resources
    Total Experience:Number of Years of experience the resource has
    Prior Industry Exposure:Prior experience in Industries
    Education:Highest Education of the Resource
    Resource Rating:Overall Rating of the resource
    The remaining columns are resource specific: 1 if the resource has that skill, else 0.

    Instructions:
    1. You have been given a new project with its solution Client Industry , domain, functional area, and required tools for the project. You would need to create a skill matrix of Resources(employees) based on the Skills, Rating, Experience and Domain.
    2. Using the skill matrix, suggest 5 resources that would best fit according to the skill matrix to match the project requirements. Take into consideration the Bandwith available for each resource. Provide justification for the choices also.

    Input format: You would be given a JSON text in the format:
    {json.dumps(usr_fmt, indent=2)}

    Input: {usr_pmt}

    Task: Assign TOP 5 resources accurately to the Project requirements based on Resources skills, experience, rating, domain and availability in the format:
    Output format:
    Project Name: <PROJECT NAME>
    Resource Name: <RESOURCE NAME>
    Designation: <PRESENT DESIGNATION>
    Skill: <RESOURCE SKILL>
    Reason: <REASON TO SELECT THE RESOURCE>
    

    ### END OF DATA ###
    """

    return prompt
```
